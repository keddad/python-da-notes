```python
import pandas as pd
```

```python
df = pd.read_csv('citibike.csv')
```

Датафрейм можно агрегировать по некоторому столбцу, используя метод groupby. Группы представляются словарем, где группируемому значению сопоставлены индексы строк

```python
df.groupby(['usertype']).groups
```

С помощью метода first можно вывести первую строку из каждых групп

```python
df.groupby(['bikeid']).first()
```

Аггрегация чаще всего выполняется, чтобы вычислить что-то по группе, например среднюю продолжительность поездок:

```python
df.groupby(['bikeid'])['tripduration'].mean()
```

Либо найти количество велосипедов на станциях

```python
df.groupby(['start station name'])['bikeid'].count()
```

Группировка может происходить по нескольким столбцам

```python
df.groupby(['start station name', 'end station name'])['tripduration'].mean()
```

#### Выходной формат groupby - series или dataframe

```python
df.groupby(['bikeid'])['tripduration'].mean().head(3)
```

```python
df.groupby(['bikeid'])[['tripduration']].mean().head(3)
```

```python
df.groupby('starttime', as_index=False).agg({"tripduration": "sum"})
```

#### Функция agg 

```python
df.groupby(['start station name']).agg({'tripduration': sum, 
                                     'usertype': "count", 
                                     'starttime': 'first'}) 
```

Можно определить функции для агрегации в словаре и передать в agg

```python
aggregations = {'tripduration': lambda x: max(x), 
                'usertype': "count", 
                'starttime': 'first'
               }
```

```python
df.groupby('bikeid').agg(aggregations).head()
```

Если необходимо после агрегации найти среднее, минимальное и максимальное значение, то эти функции можно указать в листе

```python
df.groupby(['start station name', 'bikeid']).agg({'tripduration': [min, max, sum],
                                     'starttime': [min, 'first', 'nunique']}) 
```
